## JFET

```python
import numpy as np
import matplotlib.pyplot as plt

# Computational Parameters
z = -0.4
x = np.linspace(0, 1, 100)

# P-Theory (y = IDsat / ID0)
Numerator = x - 1 - (2 / 3) * (z - 1) * (1 - ((z - x) / (z - 1)) ** 1.5)
Denominator = -1 - (2 / 3) * (z - 1) * (1 - (z / (z - 1)) ** 1.5)
yP = Numerator / Denominator

# Square-Law Theory
yS = (1 - x) ** 2

# Plot
plt.plot(x, yP, label='IDsat equation')
plt.plot(x, yS, '--', label='Square-Law')
plt.grid(True)
plt.xlabel('VG/VP')
plt.ylabel('IDsat/ID0')
plt.text(0.38, 0.4, 'Square-Law')
plt.text(0.27, 0.2, 'IDsat equation')
plt.text(0.8, 0.83, 'Vbi/VP = -0.4')

# Display the plot
plt.legend()
plt.show()
```
