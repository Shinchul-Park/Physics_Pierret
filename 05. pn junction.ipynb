```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
T = 300  # Temperature in Kelvin
k = 8.617e-5  # Boltzmann constant in eV/K
e0 = 8.85e-14  # Permittivity of free space (F/cm)
q = 1.602e-19  # Elementary charge (C)
KS = 11.8  # Relative permittivity of Si
ni = 1e10  # Intrinsic carrier concentration (cm^-3)
EG = 1.12  # Bandgap energy (eV)

# User input for doping concentrations
#NA = float(input("Please enter p-side doping (cm^-3), NA = "))
#ND = float(input("Please enter n-side doping (cm^-3), ND = "))
NA=1e16
ND=1e15

# Built-in potential calculation
Vbi = k * T * np.log((NA * ND) / ni**2)

# Depletion widths
xN = np.sqrt(2 * KS * e0 / q * NA * Vbi / (ND * (NA + ND)))
xP = np.sqrt(2 * KS * e0 / q * ND * Vbi / (NA * (NA + ND)))

# Spatial range for calculations
xleft, xright = -3.5e-4, 3.5e-4
x = np.linspace(xleft, xright, 501)

# Electrostatic potential (Vx)
Vx1 = (Vbi - q * ND * (xN - x)**2 / (2 * KS * e0) * (x <= xN)) * (x >= 0)
Vx2 = 0.5 * q * NA * (xP + x)**2 / (KS * e0) * np.logical_and(x >= -xP, x < 0)
Vx = Vx1 + Vx2

# Maximum voltage for plotting
Vmax = 3
EF = Vx[0] + Vmax / 2 - k * T * np.log(NA / ni)

# Electric field (E)
EX = -np.diff(Vx) / np.diff(x)

# Charge density (ρ)
PX = KS * e0 * np.diff(EX) / np.diff(x[:-1])

# Plotting
plt.figure(figsize=(6, 14))

# Energy Band Diagram
plt.subplot(4, 1, 1)
plt.plot(x, -Vx + EG / 2 + Vmax / 2, label="Ec")
plt.plot(x, -Vx - EG / 2 + Vmax / 2, label="Ev")
plt.plot(x, -Vx + Vmax / 2, "w:", label="Ei")
plt.axhline(EF, color="k", linestyle="-", label="EF")
plt.axvline(0, color="g", linestyle="--", label="x=0")
plt.xlim(xleft, xright)
plt.ylim(0, Vmax)
plt.title("Band Diagram at Equilibrium")
plt.legend()
plt.grid()

# Electrostatic Potential
plt.subplot(4, 1, 2)
plt.plot(x, Vx)
plt.ylabel("V(x)")
plt.xlabel("x")
plt.title("Electrostatic Potential")
plt.grid()

# Electric Field
plt.subplot(4, 1, 3)
plt.plot(x[:-1], EX)
plt.ylabel("E(x)")
plt.xlabel("x")
plt.title("Electric Field")
plt.grid()

# Charge Density
plt.subplot(4, 1, 4)
plt.plot(x[:-2], PX)
plt.ylabel("ρ(x)")
plt.xlabel("x")
plt.title("Charge Density")
plt.grid()

plt.tight_layout()
plt.show()
```
